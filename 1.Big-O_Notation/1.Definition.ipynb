# Big-O Notation

BigO notation is a mathematical notation that describes the limiting behavior of a function when the arguments tends towards a particular value or infinity.

## Formal definition
$f(n) = O(g(n))$, if there exists positive constant $c$, and $n_{o}$ such that $f(n) \le c*g(n)$ for all $n \ge n_{o}$

### Math example

$f(n) = 4n^{2} + 16n +2$

Is $f(n) => O(n^{4})$ ?

Let's assume $c=1$

| n | $4n^{2} + 16n +2 < n^{4}$ | ?   |
|:-:|:-------------------------:|-----|
| 0 |          $2 < 0$          | No  |
| 1 |         $22 < 1$          | No  |
| 2 |         $50 < 0$          | No  |
| 3 |         $86 < 81$         | No  |
| 4 |       $130 < 256$         | Yes |

So is $f(n) => O(n^{4})$ ?
For constant $c=1$ and $n_{o} = 4$ yes.

If we could never find a constatnt and starting value of $n_{o}$ from which it's become true forever the answer would be no.

More math examples: [Big O Notation (and Omega and Theta)](https://youtu.be/ei-A_wy5Yxw)

---

In plain words it's a way of analyzing how fast a program runtime grows asymptotically. That is, as the size of your input increases towards infinity.

### Runtimes of common Big-O functions

| $Big-O$        | Name        |
|----------------|-------------|
| $O(1)$         | Constant    |
| $O(\log(n))$   | Logarithmic |
| $O(n)$         | Linear      |
| $O(n*\log(n))$ | Log Linear  |
| $O(n^{2})$     | Quadratic   |
| $O(n^{3})$     | Cubic       |
| $O(2^{n})$     | Exponential |

![Big-O complexity chart](https://github.com/rszamszur/google-interview-preparation/blob/master/assets/images/big-o_complexity_chart.png?raw=true)


```python
import numpy as np
import matplotlib.pyplot as plt


# Stylesheets defined in Matplotlib
plt.style.use('bmh')

# Set up runtime comparisons
n = np.linspace(1, 10)
big_o = {
    "O(1)": np.ones(n.shape),
    "O(log n)": np.log(n),
    "O(n)": n,
    "O(n log n)": n * np.log(n),
    "O(n^2)": n ** 2,
    "O(n^3)": n ** 3,
    "O(2^n)": 2 ** n,
}

# Plot setup
plt.figure(figsize=(12, 10))
plt.ylim(0, 50)

for key, value in big_o.items():
    plt.plot(n, value, label=key)

plt.legend(loc=0)
plt.ylabel('Relative Runtime')
plt.xlabel('Input Size')
```
